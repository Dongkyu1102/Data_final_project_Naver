# 열(컬럼) 명칭 통일

```python
new_column_names = ['content', 'score', 'date', 'option', 'satis_list', 'brand', '한달사용기', '재구매', '가격', '세정력/거품', '향기', '트러블/진정', '제품구성', '배송/서비스/고객대응', '탈모 효과 및 쿨링감', '머릿결*']
df_1.columns = new_column_names
df_2.columns = new_column_names
df_3.columns = new_column_names
df_4.columns = new_column_names
df_5.columns = new_column_names
df_6.columns = new_column_names
df_7.columns = new_column_names
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7])
df = df.reset_index(drop=False)
df = df.drop('index', axis=1)

df
```

# 레이블 모두 NuN(작업 전 데이터) 추출

```python
label = ['가격', '세정력/거품', '향기', '트러블/진정', '제품구성', '배송/서비스/고객대응', '탈모 효과 및 쿨링감', '머릿결*']
label_all_null = df.loc[df[label].isnull().all(axis=1)]
```

```python
df_train = df.drop(label_all_null.index)
df_train
```

```python
len(df) == len(label_all_null) + len(df_train)
# df: 병합 전체 데이터 / labeling_null: 모든 라벨링 Null / df_train.: 라벨링 작업한 것들
```
